#1. **Discuss string slicing and provide examples.**
   
   **String Slicing:**
     String Slicing is a technique to get specific count of results.
     with the help of strings we keep telling the program that from to start and end the slice of particular string.
     We can start slice from index 0 to the length of the string.

     **Parameters:**
     
     **Start:**
     Here we give the starting index for the slice. If we do no give any starting index, it automatically starts with zero.

     **End:**
     Here we tell where to stop. If we do not give any ending value it will automatcally itrate to the last index.

     **step:** The step size (optional). If omitted, defaults to 1, meaning each character in the specified range is taken.

    **Examples:**

      **Basic Slicing:**
      If start is omitted, it starts from the beginning of the string. If end is omitted, it goes to the end of the string.

      s = "This is string slicing."
      print(s[9:21])
      Note: Now the output will be 'string slicing'

      **Omitting start and end:**
      s = "This is string slicing."
      print(s[:10])
      print(s[17:])

      **Negative Indexing:**
      Negative indices start from the end of the string. -1 refers to the last character.

      s = "This is string slicing."
      print(s[-1:-7])
      

      **Using a Step:**
      The step value specifies the stride. In the below example, every second character is taken.

       s = "This is string slicing."
      print(s[::2])

      **Reversing a string**
      A negative step can also reverse the string.
       s = "This is string slicing."
      print(s[::-2])

  print(my_list[0])
  print(my_list[2])

  my_list = ['apple', 'banana', 'cherry', 'date', 'fig']

  For Example:

  print(my_list)

5. Describe the key features of sets and provide examples of their use.

   **Sets**
   A set is simply a collection of distinct (unique) items. We can keep various different elementsin a single set but without any duplicate value. Elements are stored in curly braces{}.

   **Features of Sets**

   **Unordered and Unindexed:**
   Sets have no order. They have no indexing, slicing or any other sequence like behaviour. When we store data in a set, it gives unordered result.

   For Example:
   s = {1, 3, 5, 7}
   print(s)  // Output might be {1, 3, 5, 7} or {7, 5, 3, 1}, order isn't guaranteed

   **Unique Elements:**
   A set automatically removes duplicate values.

   For Example:
   s = {1, 2, 2, 3}
   print(s)  // Output: {1, 2, 3}

   **Mutable:**
   Sets are mutable, we can modify, add and remove elements even after creating the sets.

   For Example:
   s = {1, 2, 3}
   s.add(4)   # Adds 4 to the set
   s.remove(2)  # Removes 2 from the set
   print(s)

   **Set Comprehension:**
   Like list comprehensions, Python supports set comprehensions to create sets based on existing iterables.
   
   For Example:
   numbers = [1, 2, 3, 4, 5, 1, 2]
   squared_set = {x ** 2 for x in numbers}
   print(squared_set)

   **Efficient Membership Testing:**
   Sets are optimized for fast membership testing, sets can check the efficiency of an item more than checking in a list or a tuple.

   For Example:
   my_set = {1, 2, 3}
   print(2 in my_set)

   **Use Cases of Sets**

   **Removing Duplicates from a List:**
   It automatically removes the duplicate values.
   
   For Example:
   l = [1, 2, 2, 3, 4, 4, 5]
   unique_elements = list(set(l))
   print(unique_elements)  

   **Efficient Membership Testing:**
   To deal with large amount of data sets we require sets to check availability of an element.

   For Example:
   big_set = {x for x in range(1000000)}
   print(999999 in big_set)

   **Mathematical Set Operations:**
   It gives common values between two or more sets.

   For Example:
   friends_A = {"John", "Jane", "Doe"}
   friends_B = {"Jane", "Doe", "Smith"}

   common_friends = friends_A & friends_B  
   print(common_friends)

   **Set Operations for Permissions or Roles:**
   With the help of sets we can give certain permissions to the user and admin.

   For Example:
   admin_permissions = {"read", "write", "delete"}
   user_permissions = {"read", "comment"}

   missing_permissions = admin_permissions - user_permissions
   print(missing_permissions)

8. Discuss the importance of  dictionary keys being immutable and provide examples.
   **Dictionary**
   In Python, it stores the data in the form of keys and values.

   For Example:
   d = {
          'name' = 'Abc'
          'Contact' = 9874563210
       }
   
   **Key Reasons for Dictionary Keys Being Immutable:**

   **Hashing and Lookup Efficiency:**
   Dictionaries use a hash table under the hood to store key-value pairs. Each key is hashed, and the hash value determines the location to store the value associated with the key.
   If the key had to be changed, its hash value may change if the key itself was changed. This would compromise the dictionaryâ€™s ability to find the key efficiently, as it relies on the key hash to retrieve it.  

   **Consistency and Integrity:**
   Immutability ensures that the key remains consistent throughout its life in the dictionary. If the key is to be changed, and changed after it is added to the dictionary, the mapping between the key and its value will not match, leading to possible unpredictable results

   For example, if you change the key, the dictionary will not know how to find the associated value because the original hash (from the unmodified key) will no longer be valid.

   **Avoiding Errors:**
   Allowing mutable keys could lead to subtle and difficult-to-debug errors. For instance, modifying a key that is already in the dictionary would make it impossible to access the associated value, but the key would still exist in the dictionary in some form.

   **Examples**

   **1. Valid Dictionary Keys: Immutable Types**
   Integers and strings are common immutable types that can be used as dictionary keys.

   For Example:
   d = {
    1: 'apple',
    'name': 'John',
    (1, 2, 3): 'coordinates'
   }
   In the example above, integers, strings, and tuples (containing immutable elements) are used as dictionary keys because they are all immutable.

   **2. Invalid Dictionary Keys: Mutable Types**
   The characters are interchangeable and cannot be used as dictionary keys. Attempting to do so raises a TypeError.

   l = [1, 2, 3]
   d = {
   l: 'invalid key'  # This will raise a TypeError
   }
   Lists can be modified spatially (e.g., to add, remove, or modify items), which makes it difficult for the dictionary to rely on the hash value of the key to store or retrieve values

   **3. Mutability**
   For Example:
   l = [1, 2, 3]
   d = {my_list: 'value'}

   // Modify the list
   l.append(4)

   // Now the dictionary's key has changed because the list was modified,
   // but the dictionary doesn't "know" that.
